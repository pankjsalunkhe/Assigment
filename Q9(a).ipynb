```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
```

```python
data = pd.read_csv("/content/Q9_a (1).csv")
data.head()
```

```python
speed = data['speed']
distance = data['dist']
```

```python
# Calculate skewness
skewness_speed = skew(speed)
skewness_distance = skew(distance)

# Calculate kurtosis
kurtosis_speed = kurtosis(speed)
kurtosis_distance = kurtosis(distance)

# Print the results
print("Skewness:")
print("Speed:", skewness_speed)
print("Distance:", skewness_distance)

print("Kurtosis:")
print("Speed:", kurtosis_speed)
print("Distance:", kurtosis_distance)
```

```python
car = pd.read_csv("/content/Q9_b (1).csv")
car.head()
```

```python

# Extract the required columns for SP and WT
SP = car['SP']
WT = car['WT']

# Calculate skewness
skewness_SP = skew(SP)
skewness_WT = skew(WT)

# Calculate kurtosis
kurtosis_SP = kurtosis(SP)
kurtosis_WT = kurtosis(WT)

# Print the results
print("Skewness:")
print("SP:", skewness_SP)
print("WT:", skewness_WT)

print("Kurtosis:")
print("SP:", kurtosis_SP)
print("WT:", kurtosis_WT)
```
