```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
car = pd.read_csv("/content/Cars (1).csv")
car
```

```python
MPG_car = car['MPG']
# Assuming you have the MPG data stored in a list or numpy array called "mpg_data"
# Calculate the mean and standard deviation of the MPG data
mean_mpg = np.mean(MPG_car)
std_mpg = np.std(MPG_car)

# Case a: P(MPG > 38)
prob_a = 1 - stats.norm.cdf(38, mean_mpg, std_mpg)

# Case b: P(MPG < 40)
prob_b = stats.norm.cdf(40, mean_mpg, std_mpg)

# Case c: P(20 < MPG < 50)
prob_c = stats.norm.cdf(50, mean_mpg, std_mpg) - stats.norm.cdf(20, mean_mpg, std_mpg)

# Print the probabilities
print("Probability (MPG > 38):", prob_a)
print("Probability (MPG < 40):", prob_b)
print("Probability (20 < MPG < 50):", prob_c)
```
